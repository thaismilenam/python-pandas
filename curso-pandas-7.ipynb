```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'pink')
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'pink')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados
```

```python
dados['Bairro'].drop_duplicates()
selecao = dados.groupby('Bairro')
```

```python
fig = selecao['Valor'].std().plot.bar(color = 'pink')
```

```python
fig = selecao['Valor'].mean().plot.bar(color = 'pink')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
